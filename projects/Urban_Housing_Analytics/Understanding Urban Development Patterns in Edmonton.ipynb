#### SECTION A: INTRODUCTION

##### Introduction
Urban development is a key indicator of any city's economic growth and population dynamics. In Edmonton, tracking construction activity over time can reveal patterns in residential expansion, commercial infrastructure development, and shifts in zoning practices. This project uses building permit data issued by the City of Edmonton since 2009 to explore the landscape of construction activity. By analyzing trends in building types, permit frequency, geographic concentration, and estimated construction value, this project aims to explore the relationship between the physical and economic growth. 

##### Background
Over the past decade, Edmonton has adopted a variety of policy initiatives to help promote higher-density development, address housing affordability, and encourage redevelopment in older neighbourhoods. For eample, the 2020 City Plan (City of Edmonton, 2020) outlines Edmonton’s growth strategy, emphasizing infill in established areas and diversifying housing types, particularly through zoning reforms like the recent city-wide Zoning Bylaw (City of Edmonton, 2024) which came into effect on January 1, 2024. These initiatives aim to accommodate future population growth.

While Edmonton’s home prices are generally lower than in other major Canadian cities, rising construction costs, and shifts in household income have led to growing disparities in access to affordable housing. Infill development and multi-family housing projects have been concentrated in some neighbourhoods, raising questions concerning gentrification and equitable investment (Edmonton Social Planning Council, 2022).

It is not clear if the impacts of these development trends are evenly distributed across the city. Some neighbourhoods have seen significant new construction and rising property values, while others have remained relatively unchanged. Without access to comprehensive data on rents or household income, it can be difficult to fully assess affordability pressures or displacement. Available data on building permits and property assessments offer valuable insight into where reinvestment is taking place, how property values are evolving, and what kinds of development are reshaping different parts of the city. This raises the questions:
- Which neighbourhoods are attracting the most investment?
- Is redevelopment contributing to rising property values—and for whom?
- And how are older neighbourhoods changing in comparison to Edmonton’s newer suburbs?

##### Objective

This analysis seeks to explore how construction activity, measured by permit counts and construction value, reflects larger patterns in urban development, zoning policy, and rising property values in Edmonton over the past 16 years.The goal is to identify whether neighbourhoods are experiencing equitable investment and development, and how these trends relate to neighbourhood and property values. These insights are valuable for city planners, developers, researchers, and residents interested in how Edmonton's urban environment is changing.

#### SECTION B: ANALYSIS QUESTIONS

The analysis in this report will be geared towards addressing the following guiding questions: 

1. Which neighbourhoods have seen the most investment since 2009? 

	This helps identify areas that have attracted the greatest construction investment over time. It offers insight into investment trends and could potentially reveals priorities in urban planning, private sector interest, or other factors.

2. How has the total value of building permits changed over time in the top 10 neighbourhoods that have received the most investment since 2009 and are there periods of increased investment in particular areas? 

	This will provide a view of changes in urban development, and could be used to identify responses to policy shifts (e.g., post-2020 City Plan).

3. Which neighborhoods are experiencing higher-value building permit activity, and how does this relate to the age or maturity of the neighborhood?

	This explores where reinvestment is occurring, particularly in older areas, and may indicate targeted redevelopment or renewal trends.

4. Is there a relationship between the volume or value of building permits and changes in assessed property values across neighbourhoods? 

	This question will help evaluate whether increased construction activity is associated with rising property values which could be interpreted as a potential signal of gentrification or market pressure.

5. What types of construction (e.g., new builds, additions, renovations) are most common in different neighborhoods, and how do their construction values vary? 

	This question will help distinguish between types of reinvestment and reflect different levels of redevelopment or maintenance.

#### SECTION C: DATA SOURCING AND JUSTIFICATION

##### Dataset Sourcing and Justification
The datasets used in this analysis are:
- The building permits dataset from the City of Edmonton - Urban Planning & Economy Department for the public to do construction or maintenance on a structure located in the City, with records starting from January 1, 2009. 
- The Property Assessment Data provided by the City Administration. This dataset contains the historical assessed value of properties within the City of Edmonton from 2012-01-01 until 2024-12-31 but was filtered to records from the last 5 years(2019 to 2024) due to its size.

The datasets were sourced from the City of Edmonton's Open data portal and is under the City of Edmonton's Open Government Licence, which makes it available for use in this context. It can be accessed through these links : [permits](https://data.edmonton.ca/Urban-Planning-Economy/General-Building-Permits/24uj-dj8v/about_data), [property assessments](https://data.edmonton.ca/City-Administration/Property-Assessment-Data-Historical-/qi6a-xuwt/about_data)

Contains information licensed under the [Open Government Licence – City of Edmonton](https://data.edmonton.ca/stories/s/Open-Data-Terms-of-Use/bdmk-rzti/).

This analysis is based on publicly available data and does not imply any endorsement by the City of Edmonton. The data is provided “as is” without warranty of any kind.

These datasets were selected from the City of Edmonton's Open Data Portal due to their comprehensiveness and direct connection to urban development trends. The permits dataset contains over 224,000 records and 23 attributes such as permit date, job and building type, construction value, and geospatial location, making it well-suited for both temporal and spatial analysis. The property assessments dataset contained slightly over 2 million records after filtering, and contains 21 attributes in total, but the attributes of interest include the Assessment year, Neighbourhood, Zoning and Assessed Value. The two datasets will be combined by grouping records by neighbourhood and assessmnet year in the property assessments dataset, summing them up to compute the total property value assessments for each year and then performing a join based on the neighbourhood and yeqr. 

These datasets are suitable for addressing the project’s objective because the data it captures enables longitudinal analyses of construction patterns, comparisons across neighbourhoods, and correlations between variables like property value and construction cost. Its text, numeric, and geospatial fields allow for diverse visualization techniques, including maps, time series, and plots of distributions.

##### 1. Which neighbourhoods have seen the most investment since 2009? 
Using a horizontal bar chart, the Top 10 Neighbourhoods by total construction activity since 2009 can be visualizaed to show which neighbourhoods have seen the most construction activity. The bar chart makes it easy to compare relative activity across neighbourhoods.

Windermere by far is the leading neighbourhood in terms of construction investment with a accumulated value of $1.72 billion, followed by Chappelle with $1.42 billion
and Laurel with $1.25 billion. These three neighbourhoods have each received over $1.2 billion in building permit value since 2009.

Other notable high-investment areas are Summerside, McConachie, and Walker. Each around having about $1.23 billion in investment as well as Secord and Keswick which have just over $1 billion.

All of these are newer suburban or developing areas, reflecting Edmonton’s trend toward peripheral urban growth and greenfield development.

Since 2009, the most investment has concentrated in Windermere, Chappelle, and Laurel, with all top 10 neighborhoods being new or developing suburbs. No mature or central neighborhoods like Downtown appear in this list, highlighting a pattern of suburban growth over urban reinvestment.

This suggests a strong expansion-driven development strategy over the past 15 years, with limited focus on redevelopment in older or core areas.

##### 2. How has the total value of building permits changed over time in the top 10 neighbourhoods that have received the most investment since 2009?

A line chart is ideal for visualizing trends over time such as changes and fluctuations in value year-over-year, and comparing multiple categories (neighbourhoods in this case) across time. Other charts like bar or area charts could get cluttered when comparing multiple time series, making them more difficult to read

This line chart tracks yearly construction investment trends from 2009 to 2024 in Edmonton’s top 10 neighbourhoods by total building permit value. It depicts how investment has fluctuated over time in these high-activity areas.

Windermere saw a dramatic spike in construction value around 2016, approaching $1 billion.

In almost every year from 2009 to 2025, the construction value in Not Mature neighborhoods (blue bars) is significantly higher than in Mature neighborhoods (red bars). While Mature neighborhoods do receive some investment, about $351M in 2017, and $293M in 2014, it is consistently dwarfed by the investment in younger neighbourhoods. This indicates that reinvestment in older areas is occurring, but at a much smaller scale.

There is no strong upward trend in the reinvestment into Mature neighborhoods. This could imply a lack of targeted redevelopment or renewal strategies in older areas over the 16-year span.

The data shows that higher-value building permit activity is overwhelmingly concentrated in newer or developing neighborhoods. While some reinvestment is happening in older areas, it remains smaller in comparison. These trends suggest that urban growth strategies and investment priorities continue to prioritize expansion into new developments rather than the revitalization of older neighborhoods. ##### 5. What types of construction (e.g., new builds, additions, renovations) are most common in different neighborhoods, and how do their construction values vary?

A heatmap makes it easy to compare both the magnitude and distribution of construction value across multiple categories and neighbourhoods at once using color intensity.

  <figcaption><strong>Figure 1:</strong> Common types of construction work by neighbourhood.</figcaption>

The heatmap shows that different neighbourhoods specialize in different types of construction.

Commercial–Final projects tend to dominate in core and mixed‑use areas like Downtown, Griesbach, and Chappelle, indicating a focus on finishing commercial developments in those parts of the city.

Home Improvement seems to be more common in mature residential areas such as Summerside, The Hamptons, and Walker, suggesting that homeowners in these established communities are investing heavily in renovations.

House Combination and Accessory Building Combination values are highest in newer areas like Laurel and Rosenthal, which could suggest secondary-suite or backyard‑suite growth in those neighbourhoods.

Single‑semi‑detached & rowhousing appears to be the strongest in growth areas like Windermere and Secord, reflecting the rise of medium‑density residential builds there.

Overall, the pattern shows that commercial zones and infill areas focus on commercial and new residential construction, while older suburbs concentrate on home improvements. These insights can guide planners and developers to target neighbourhood‑specific opportunities rather than apply a one‑size‑fits‑all approach.
